```python
import os
from pathlib import Path

from dotenv import load_dotenv
from tqdm import tqdm

from chase_rank.wrappers import (
    ValhallaHandler, TrackHandler, MatchHandler
)
```

```python
nord7 = "#8fbcbb" # frost
nord8 = "#88c0d0" # frost
nord9 = "#81a1c1" # frost
nord10 = "#5e81ac" # frost
nord11 = "#bf616a" # aurora
nord12 = "#d08770" # aurora
nord13 = "#ebcb8b" # aurora
nord14 = "#a3be8c" # aurora
nord15 = "#b48ead" # aurora

colors = [nord11, nord12, nord13, nord14, nord7, nord8, nord9, nord10]
```

```python
load_dotenv("../.env")

TRACK_PATH = Path(os.getenv("TRACK_PATH", "../data/tracks"))
MATCH_PATH = Path(os.getenv("MATCH_PATH", "../data/matches"))
```

# Initialize Handlers

```python
tracks = TrackHandler(TRACK_PATH)
matches = MatchHandler(MATCH_PATH)
valhalla = ValhallaHandler()
```

# Match Tracks

### Match single track

```python
track = tracks.get(tracks.track_id_list[0])
match = valhalla.match(track)

match.surface.value_counts(dropna=False)
```

```python
match[["geometry", "surface"]].explore("surface",cmap=colors)
```

### Try to match all activities we have tracks for but no matched tracks yet
This will also retry to match tracks we failed to match the last times

```python
unmatched_tracks = [activity_id
                    for activity_id in tracks.track_id_list
                    if activity_id not in matches.match_id_list]

for activity_id in tqdm(unmatched_tracks):
    match = valhalla.match(tracks.get(activity_id))
    if match is not None:
        matches.add_match(activity_id, match)
```
