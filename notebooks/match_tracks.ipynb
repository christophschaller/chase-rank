```python
import os
from pathlib import Path

from dotenv import load_dotenv
from tqdm import tqdm

from chase_rank.data_handlers import TrackHandler, MatchHandler
from chase_rank.service_handlers.valhalla_handler import ValhallaHandler
```

```python
load_dotenv("../.env")

USER_PATH = Path(os.getenv("USER_JSON_PATH", "../data/users.json"))
ACTIVITY_PATH = Path(os.getenv("ACTIVITY_PATH", "../data/activities.parquet"))
TRACK_PATH = Path(os.getenv("TRACK_PATH", "../data/tracks"))
MATCH_PATH = Path(os.getenv("MATCH_PATH", "../data/matches"))

STRAVA_CLIENT_ID = os.getenv("STRAVA_CLIENT_ID")
STRAVA_CLIENT_SECRET = os.getenv("STRAVA_CLIENT_SECRET")
```

# Initialize Handlers

```python
track_handler = TrackHandler(TRACK_PATH)
match_handler = MatchHandler(MATCH_PATH)
valhalla_handler = ValhallaHandler()
```

### Try to match all activities we have tracks for but no matched tracks yet
This will also retry to match tracks we failed to match the last times

```python
for activity_id in tqdm([activity_id
                         for activity_id in track_handler.tracklist
                         if activity_id not in match_handler.match_ids]):
    print(activity_id)
    match = valhalla_handler.match(track_handler[activity_id])
    if match is not None:
        match_handler.add_match(activity_id, match)
    else:
        print(f"could not match {activity_id}")
```
